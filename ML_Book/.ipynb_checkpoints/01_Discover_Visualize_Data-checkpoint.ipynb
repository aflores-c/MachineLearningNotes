## Explore the data

Import your machine learning libraries

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('housing.csv')
```

This is useful to check the firsts row of your data

```python
df.head()
```

Use df.info to check the number of columns or the data types

```python
df.info()
```

Check if there are null values

```python
df.isnull().any()
```

Count the values of a column

```python
df['diagnosis'].value_counts()
```

Get a summary of the numerical attributes. 

```python
df.describe()
```

Plot a histogram of the numerical attributes

```python
df.hist(bins=10, figsize=(20,15), grid = False)
plt.show()
```

We can see that there is too much data, so we can select the data that is important for us. Let's say we just want the first 11 columns

```python
data=df.iloc[:,1:11]
```

We can see the correlations

```python
data.corr()
```

We can plot density charts

```python
data.plot(kind= 'density', subplots=True, layout=(4,3), sharex=False, 
                     sharey=False,fontsize=12, figsize=(15,10))
plt.show()
```

We can analyze by box plots too

```python
data.plot(kind= 'box' , subplots=True, layout=(4,4), sharex=False, sharey=False,fontsize=12, figsize=(15,10))
plt.show()
```
